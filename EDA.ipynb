<div>
    <p class= "titulo2" align="center">1.  Planteamineto del problema</p>
    <p class= "subtitulo2">En la actualidad, el acceso a la educación superior es considerado un derecho clave para el desarrollo personal y profesional. Sin embargo, la eficacia con la que los estudiantes eligen sus campos de estudio y la relación entre estas elecciones y las oportunidades laborales posteriores varía significativamente, además de que el mundo está migrando a un futuro incierto con la influencia de la inteligencia artificial, la cual está trayendo un conjunto de cambios demasiado importantes y de una manera muy rápida. Así como estos,  también hay otros factores (como las universidades en línea) que permiten a los estudiantes aprender a su propio ritmo de una manera más eficiente.
    </br></br>
    Esto plantea una pregunta fundamental: ¿los sistemas de selección y orientación en la educación superior garantizan oportunidades que perduran en el tiempo y, a su vez, sean otorgadas a aquellos que las aprovecharán de manera óptima?
    </br></br>
    El problema central radica en que, en muchos casos, las decisiones sobre el acceso a estudios superiores no se basan en un análisis profundo de las aptitudes, intereses y proyecciones laborales de los aspirantes. Esto puede resultar en una mala asignación de recursos, altos índices de deserción, y graduados que terminan trabajando en ocupaciones no relacionadas con sus áreas de estudio, muchas veces con menores beneficios económicos y personales.
    </br></br>
    Por lo tanto, surge la necesidad de evaluar si el sistema actual de selección para estudios superiores está alineado con los intereses de los estudiantes y las demandas del mercado laboral.
    </br></br>
    El dataset Field of Study vs Occupation de Kaggle ofrece una valiosa oportunidad para explorar cómo los campos de estudio se correlacionan con las ocupaciones y otros indicadores laborales, como ingresos, empleabilidad y satisfacción profesional. Al realizar un Análisis Exploratorio de Datos (EDA), es posible identificar patrones clave que evidencian desajustes entre las aspiraciones educativas de los estudiantes y las demandas del mercado laboral.
    </br></br>
    A través de este análisis, se busca generar evidencia que pueda fundamentar la creación o mejora de políticas de orientación vocacional y selección, asegurando que las oportunidades educativas se brindan a quienes tengan mayores probabilidades de beneficiarse de ellas, tanto a nivel individual como social.

<div>
    <p class= "titulo2" align="center">4. Descripción del Dataset "Field of study"</p>
    </br>
    <ul class="subtitulo2">
    <li><strong>Field of Study (Campo de estudio):</strong>
        <p>Tipo: Categórica (String)</p>
        <p>Descripción: El área académica en la que se enfocó el individuo durante su educación.</p>
        <p>Importancia: Permite identificar la relación entre los campos de estudio y las ocupaciones laborales, lo cual es clave para evaluar la alineación educativa y profesional.</p>
    </li>
    <li><strong>Current Occupation (Ocupación actual):</strong>
        <p>Tipo: Categórica (String)</p>
        <p>Descripción: El trabajo o industria actual del individuo.</p>
        <p>Importancia: Ayuda a establecer patrones de transición entre campos de estudio y trabajos específicos.</p>
    </li>
    <li><strong>Age (Edad):</strong>
        <p>Tipo: Entero</p>
        <p>Descripción: La edad del individuo.</p>
        <p>Importancia: Es útil para analizar tendencias relacionadas con la experiencia y etapas de la carrera profesional.</p>
    </li>
    <li><strong>Gender (Género):</strong>
        <p>Tipo: Categórica (String)</p>
        <p>Descripción: El género del individuo.</p>
        <p>Importancia: Permite analizar disparidades de género en ocupaciones e ingresos.</p>
    </li>
    <li><strong>Years of Experience (Años de experiencia):</strong>
        <p>Tipo: Entero</p>
        <p>Descripción: Los años que el individuo ha estado en la fuerza laboral.</p>
        <p>Importancia: Es relevante para evaluar la relación entre experiencia laboral y progresión profesional.</p>
    </li>
    <li><strong>Education Level (Nivel educativo):</strong>
        <p>Tipo: Categórica (String)</p>
        <p>Descripción: El nivel más alto de educación alcanzado por el individuo.</p>
        <p>Importancia: Es clave para comprender cómo los niveles educativos afectan la empleabilidad, los ingresos y por consecuencia al cambio de especialidad u ocupación.</p>
    </li>
    <li><strong>Industry Growth Rate (Tasa de crecimiento de la industria):</strong>
        <p>Tipo: Categórica (String)</p>
        <p>Descripción: La tasa de crecimiento de la industria donde trabaja el individuo.</p>
        <p>Importancia: Ayuda a identificar oportunidades laborales en sectores en expansión.</p>
    </li>
    <li><strong>Job Satisfaction (Satisfacción laboral):</strong>
        <p>Tipo: Entero (Escala 1-10)</p>
        <p>Descripción: Calificación de la satisfacción laboral del individuo.</p>
        <p>Importancia: Permite medir la felicidad y motivación en el trabajo, influenciando la productividad.</p>
    </li>
    <li><strong>Work-Life Balance (Equilibrio trabajo-vida):</strong>
        <p>Tipo: Entero (Escala 1-10)</p>
        <p>Descripción: Calificación del equilibrio percibido entre trabajo y vida personal.</p>
        <p>Importancia: Es un indicador del bienestar laboral y personal.</p>
    </li>
    <li><strong>Job Opportunities (Oportunidades laborales):</strong>
        <p>Tipo: Entero</p>
        <p>Descripción: Número de oportunidades laborales disponibles en el campo del individuo.</p>
        <p>Importancia: Ayuda a analizar la demanda en ciertos campos ocupacionales.</p>
    </li>
    <li><strong>Salary (Salario):</strong>
        <p>Tipo: Entero</p>
        <p>Descripción: El salario anual del individuo (en USD o moneda local equivalente).</p>
        <p>Importancia: Es crucial para evaluar la relación entre estudios, ocupación e ingresos.</p>
    </li>
    <li><strong>Job Security (Seguridad laboral):</strong>
        <p>Tipo: Entero (Escala 1-10)</p>
        <p>Descripción: Calificación de la seguridad laboral percibida por el individuo.</p>
        <p>Importancia: Influye en la estabilidad y satisfacción laboral del individuo.</p>
    </li>
    <li><strong>Career Change Interest (Interés en cambio de carrera):</strong>
        <p>Tipo: Entero (0 o 1)</p>
        <p>Descripción: Indica si el individuo está interesado en cambiar de ocupación (1 para sí, 0 para no).</p>
        <p>Importancia: Mide la intención de cambio profesional, útil para estudiar las razones detrás de este fenómeno.</p>
    </li>
    <li><strong>Skills Gap (Brecha de habilidades):</strong>
        <p>Tipo: Entero (Escala 1-10)</p>
        <p>Descripción: Una medida de cuán bien se ajustan las habilidades actuales del individuo a los requisitos de su trabajo.</p>
        <p>Importancia: Es fundamental para identificar necesidades de capacitación.</p>
    </li>
</ul>
    </br></br>
</ul>

```python
# Ahora graficamos para visualizar como estan distribuidos los datos entre las profesiones seleccionadas
import matplotlib.pyplot as plt
import numpy as np

field = dataset['Field of Study'].value_counts()

plt.bar(field.index, field.values, width=0.7, edgecolor="blue", linewidth=0.7)
# Etiquetas y título
plt.title("Gráfico Profesiones")
plt.xlabel("Profesiones")
plt.ylabel("Encuestados")
plt.xticks(ticks=field.index, labels=field.index)

# Mostrar el gráfico
plt.show()
```

```python
porcentajes = []
for i in range(10):
    porcentajes.append(float(field[i] * 100 / field.sum())) 
suma = pd.Series(porcentajes, index=field.index)
suma
```
